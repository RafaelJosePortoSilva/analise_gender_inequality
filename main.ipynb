```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

dados = df.iloc[110,8:]

plot_acf(dados, lags=len(dados)-1)
plt.title('Função de Autocorrelação (ACF)')
plt.show()

plot_pacf(dados, lags=15)
plt.title('Função de Autocorrelação Parcial (PACF)')
plt.show()
```

```python
dt = np.asarray(dados).astype('float64')
```

```python
from statsmodels.tsa.arima.model import ARIMA
mod = ARIMA(endog=dt,
            order=(5,1,3))
res = mod.fit()
```

```python
import statsmodels as st
prev = res.get_forecast(steps=10)
prev_usar = prev.predicted_mean
```

```python
graph_data = np.concatenate([dt,prev_usar])
#np.asarray(graph_data)
graph_data
```

```python
sns.lineplot(data=np.asarray(graph_data))
```
