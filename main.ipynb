```python
import pandas as pd
import numpy as np
```

```python
origin = pd.read_csv('Gender Inequality Index.csv')
df = origin.copy()
```

```python
df.columns = list(map(lambda x: x.replace('Gender Inequality Index ','gii'),np.asarray(df.columns)))
```

```python
df.columns = list(map(lambda x: x.lower(),np.asarray(df.columns)))
```

```python
df.columns = list(map(lambda x: x.replace(' ','_'),np.asarray(df.columns)))
```

```python
df.columns
```

```python
df.rename(columns={
    'human_development_groups':'hdg',
    'undp_developing_regions':'undp_dr',
    'hdi_rank_(2021)':'hdi_rank',
    'gii_rank_(2021)':'gii_rank'
},inplace=True)
```
